```python
def csv():
    """
    on importe le csv   
    """
  
file = 'C:/Users/evirevialle/persons.csv'
data = pd.read_csv(file, sep = ',')
columns_number=data.shape[1]
noble_number=data.shape[0]
data
```

```python
def carte():
    """
    On va créer une carte avec le package folium
    """

m = folium.Map(
    location=[48.8566969, 2.3514616],
    zoom_start = 12,
    control_scale = True,
    max_bounds=True,
    max_lat=48.813141, max_lon=2.234129,
    min_lat=48.908715, min_lon=2.422941,
)
m.fit_bounds([[48.813141, 2.234129],[48.908715,2.422941]])

m
```

```python
def markers():
    """
    On rajoute les marqueurs en créant une boucle
  """

featuregroup1 = folium.FeatureGroup(name = 'F')
featuregroup2 = folium.FeatureGroup(name = 'M', show=False)
m.add_child(featuregroup1)
m.add_child(featuregroup2)
cluster = folium.plugins.MarkerCluster().add_to(featuregroup1)
cluster2 = folium.plugins.MarkerCluster().add_to(featuregroup2)
folium.LayerControl().add_to(m)


for i in range(data.shape[0]):
    try:
        location = [data.groupby('gender').get_group('F')['geo_lat'][i], data.groupby('gender').get_group('F')['geo_long'][i]]
        folium.Marker(location).add_to(cluster)
        locationM = [data.groupby('gender').get_group('M')['geo_lat'][i], data.groupby('gender').get_group('M')['geo_long'][i]]
        folium.Marker(locationM).add_to(cluster2)
    except KeyError:
        True

m
```
